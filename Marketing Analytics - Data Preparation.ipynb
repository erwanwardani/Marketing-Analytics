```python
# Import Library
import pandas as pd
import numpy as np
```

```python
# Read Data
data = pd.read_csv("final_shop_6modata.csv")

# Show Data
data
```

```python
# Show Information of Data
data.info()
```

```python
# Missing Value Check
data.isnull().sum()
```

```python
# Duplicate Check
data.duplicated().sum()
```

```python
# Casting Data Type
data["Ad Group"] = data["Ad Group"].astype("string")

# Show Updated Information of Data
data.info()
```

```python
# Extract Ad Type Data
data['Ad Type'] = data['Ad Group']

data['Ad Type'][data['Ad Type'].str.contains('discount')] ='Discount'
data['Ad Type'][data['Ad Type'].str.contains('Discount')] ='Discount'
data['Ad Type'][data['Ad Type'].str.contains('promo')] ='Promo'
data['Ad Type'][data['Ad Type'].str.contains('Promo')] ='Promo'
data['Ad Type'][data['Ad Type'].str.contains('coupon')] ='Coupon'
data['Ad Type'][data['Ad Type'].str.contains('Coupon')] ='Coupon'
data['Ad Type'][data['Ad Type'].str.contains('Competitor')] ='Competitor'
data['Ad Type'][data['Ad Type'].str.contains('Sale')] ='Sale'
data['Ad Type'][data['Ad Type'].str.contains('Black Friday')]='Black Friday'
data['Ad Type'][data['Ad Type'].str.contains('Free Shipping')] ='Free Shipping'
data['Ad Type'][data['Ad Type'].str.contains('Offer')] ='Offer'

# Extract Device Data
data['Device'] = data['Ad Group']

data['Device'][data['Device'].str.contains('Desk')] ='Desktop'
data['Device'][data['Device'].str.contains('Mob')] ='Mobile'

# Extract Marketing Method Data
data['Strategy'] = data['Ad Group']

data['Strategy'][data['Strategy'].str.contains('1:1')] ='Personalized'
data['Strategy'][data['Strategy'].str.contains('Exact')] ='Exact Match'
data['Strategy'][data['Strategy'].str.contains('Phrase')] ='Phrase Match'

# Show Extracted Data
data[['Ad Group', 'Ad Type', 'Device', 'Strategy']]
```

```python
# Calculate Cost Per Mile (CPM) Metric
data['CPM'] = (data['Cost'] / data['Impressions']*1000).round(2)

# Show Calculated CPM
data[['Impressions', 'Cost', 'CPM']].head()
```

```python
# Show Information of Data
data.info()
```

```python
# Show Updated Data
data
```

```python
# Export Data to .csv
data.to_csv("final_shop_6modat_clean.csv", index=False)
```
